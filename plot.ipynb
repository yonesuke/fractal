```python
from fractals import barnsley_fern, hilbert_curve, sierpinski_triangle, dragon_curve
import matplotlib.pyplot as plt
import os
```

```python
dir_fig = 'figures'
```

```python
# barnsley fern
n_points = 100000
x, y = barnsley_fern(n_points)
plt.figure(figsize=(10, 10))
plt.scatter(x, y, s=0.2, c='green')
plt.savefig(os.path.join(dir_fig, 'barnsley_fern.png'))
```

```python
# hilbert curve
n_iter = 7
x, y = hilbert_curve(n_iter)
plt.figure(figsize=(10, 10))
plt.axis('equal')
plt.plot(x, y, linewidth=0.5, c='black')
plt.savefig(os.path.join(dir_fig, 'hilbert_curve.png'))
```

```python
# sierpinski triangle
n_iter = 7
triangles = sierpinski_triangle(n_iter)
plt.figure(figsize=(10, 10))
plt.axis('equal')
for (x, y) in triangles:
    plt.plot(x, y, linewidth=0.5, c='b')
plt.savefig(os.path.join(dir_fig, 'sierpinski_triangle.png'))
```

```python
# dragon curve
n_iter = 15
x, y = dragon_curve(n_iter)
plt.figure(figsize=(10, 10))
plt.axis('equal')
plt.plot(x, y, linewidth=0.5, c='y')
plt.savefig(os.path.join(dir_fig, 'dragon_curve.png'))
```
